```python
path = '/content/Screenshot 2024-12-03 142351.png'
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageFilter
from scipy.ndimage.filters import convolve
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
sharpen = np.array([

    [0,-2,0],

    [-2,11,-2],

    [0,-2,0]

])
```

```python
im = np.array(img,dtype=float)
kernel = sharpen
im2=convolve(im,kernel)
im3=np.array(np.clip(im2,0,255),dtype=np.uint8)
cv2_imshow(im3)
```

```python
img_new=cv2.hconcat([img,im3])
cv2_imshow(img_new)
blur1 = np.ones((7,7),dtype=float)*(1/49)
```
