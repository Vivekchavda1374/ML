```python
import pandas as pd
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ["Sepal_Length" , "Sepal_Width" , "Petal_Length" , "Petal_Width" , "Class"]
dataset = pd.read_csv(url , names = names)
dataset
```

```python
traningclass = dataset.values[: , -1]
traningclass
```

```python
unique_list = list(set(traningclass))
unique_list
```

```python
for i in range(len(traningclass)) :
  for j in range(len(unique_list)) :
    if traningclass[i] == unique_list[j] :
      traningclass[i] = j
```

```python
traningclass = traningclass[:-1]
```

```python
def Euclidian_Distance(rows_i , rows_j) :
  distance = 0.0

  for i in range(len(rows_i)) :
    distance += (rows_i[i] - rows_j[i])**2
  return np.sqrt(distance)
```

```python
training = dataset.values[: , :-1]
training
```

```python
testing = dataset.values[149 , :-1]
training = dataset.values[:149 , :-1]
```

```python
distance = []

for i in range(len(training)) :
  distance.append([Euclidian_Distance(training[i] , testing) , traningclass[i]])
```

```python
distance.sort()
distance
```

```python
k = 25

result = {}

for i in range(k) :
  result[distance[i][1]] = result.get(distance[i][1] , 0) + 1
```

```python
max_key = max(result, key=result.get)
max_key
```

```python
class_name = unique_list[max_key]
class_name
```
